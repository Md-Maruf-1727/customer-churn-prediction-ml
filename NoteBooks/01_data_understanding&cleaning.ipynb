```python
df = df.drop(['customerID'], axis=1)
```

```python
#Customer Personal Info
sel_p = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']
for col in sel_p:
    print(df[col].value_counts())
    print("--")
```

```python
# Additional Services
sel_a = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection']
for col in sel_a:
    print(df[col].value_counts())
    print("--")
```

```python
# Contract & Billing Info
sel_i = ['Contract', 'PaperlessBilling', 'PaymentMethod']
for col in sel_i:
    print(df[col].value_counts())
    print("--")
```

```python
# Target
plt.figure(figsize=(4,3))
sns.countplot(df['Churn'], color="#28CFCF", edgecolor='black')
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].isna().sum()
```

```python
df = df.dropna()
```

```python
df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')
```

```python
sel_col =[
    'OnlineSecurity', 
    'OnlineBackup', 
    'DeviceProtection', 
    'TechSupport',
    'StreamingTV', 
    'StreamingMovies'
]
for col in sel_col:
    df[col] = df[col].replace('No internet service', 'No')
```

```python
sel_col = ['tenure', 'MonthlyCharges', 'TotalCharges']
fig, axes = plt.subplots(1, 3, figsize=(15,3))
for ax, col2 in zip(axes.flatten(), sel_col):
    sns.boxplot(
        x=df[col2],
        ax = ax,
        color = "#16E6CA",
        linewidth = 1,
        linecolor = 'black',
        width=0.3,
        notch=True
    )
    ax.set_xlabel(col2, fontweight='bold')
    ax.grid(True, linestyle="-")
plt.tight_layout()
plt.show()
```
